```python
import pandas as pd
import numpy as np
import os
```

```python
DF_RAW = pd.read_csv('/workspace/nabang1010/STEAM/LeGiaBach_STEAM/src/yolov5/submission_yolov5_15.csv')
DF_RAW
```

```python
list_image = DF_RAW['ID'].unique()
len(list_image)
```

```python
df_submission = pd.DataFrame(columns=['image_id', 'PredictionString'])
for img_file in list_image:
    df_tmp = DF_RAW[DF_RAW['ID'] == img_file]
    predict_string = ''
    for j in range(len(df_tmp)):
        if df_tmp.iloc[j, :]["class_id"] == 14 and len(df_tmp) == 1:
            predict_string = predict_string + '14 1.0 0 0 1 1'
            break
        elif df_tmp.iloc[j, :]["class_id"] == 15 and len(df_tmp) == 1:
            predict_string = predict_string + '14 1.0 0 0 1 1'
            break
        else:
            if df_tmp.iloc[j, :]["class_id"] != 14 and df_tmp.iloc[j, :]["class_id"] != 15:
                predict_string += str(df_tmp.iloc[j, :]["class_id"]) + ' ' + str(df_tmp.iloc[j, :]["score"]) + ' ' + str(df_tmp.iloc[j, :]["x_min"]) + ' ' + str(df_tmp.iloc[j, :]["ymin"]) + ' ' + str(df_tmp.iloc[j, :]["x_max"]) + ' ' + str(df_tmp.iloc[j, :]["ymax"]) + ' '
    df_submission = df_submission.append({'image_id': img_file, 'PredictionString': predict_string}, ignore_index=True)
            
```
