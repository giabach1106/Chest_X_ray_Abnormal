```python
import os
import glob2
import math
import numpy as np
```

```python
files = []
for ext in ["*.png", "*.jpeg", "*.jpg"]:
  image_files = glob2.glob(os.path.join("/workspace/nabang1010/STEAM/LeGiaBach_STEAM/DATA/dataset/yolo_15", ext))
  files += image_files

nb_val = math.floor(len(files)*0.2)
rand_idx = np.random.randint(0, len(files), nb_val)

with open("/workspace/nabang1010/STEAM/LeGiaBach_STEAM/src/yolov5/val.txt", "w") as f:
    for idx in np.arange(len(files)):
        if (idx in rand_idx) and (os.path.exists(files[idx][:-3] + "txt")):
            f.write(files[idx]+'\n')

with open("/workspace/nabang1010/STEAM/LeGiaBach_STEAM/src/yolov5/train.txt", "w") as f:
  for idx in np.arange(len(files)):
    if (os.path.exists(files[idx][:-3] + "txt") and idx not in rand_idx):
      f.write(files[idx]+'\n')          
```
